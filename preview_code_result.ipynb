```python
print("1.First five Records :")
print(df.head())
print("2.Last five Records :")
print(df.tail())
print("3.Size of the dataset :", df.size)
print("\n4.Dataset shape :", df.shape)
print(df.shape[0])
print(df.shape[1])
print("\n5.List of coloumns :", df.columns)
print("\n6. Axes:", df.axes)
```
